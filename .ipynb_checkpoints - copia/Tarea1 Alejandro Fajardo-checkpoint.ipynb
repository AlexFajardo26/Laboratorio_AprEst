```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

## Tarea 1

Crea una regresión lineal con los demás inputs numericos

'housing_median_age', 'total_rooms',
'total_bedrooms', 'population', 'households'

- Escribe la ecuación del modelo

Responde la pregunta: ¿Cuál es el mejor modelo de estos? ¿Por qué?

### Datos

```python
df = pd.read_csv('housing.csv')
df
```

```python
df = df[df.median_house_value != df.median_house_value.max()]
```

### Visualización

```python
# Grafica sin los máximos
plt.scatter( df.median_income, df.median_house_value)
```

## Housing Median Age

### Análisis bivariado Housing Median Age y Median House Value

```python
df['housing_median_age_bin'] = pd.qcut(df['housing_median_age'] , q = 10, labels = False) + 1
grouped_age = df.groupby('housing_median_age_bin').mean(numeric_only = True)
plt.plot(grouped_age.housing_median_age.values, grouped_age.median_house_value.values, marker = "o")
```

#### Análisis

Podemos ver que esta relacion es cero lineal. Se puede observar claramanete como hay varias subidas y bajadas en la gráfica lo que nos expresa que no hay linealidad en esta relación, hacindo de este un modelo muy malo y poco util para poder usarloen nuestra aplicación.

### Regresión 

```python
y = df['median_house_value']
```

```python
X_age = df[['housing_median_age']]
model_age = linear_model.LinearRegression().fit(X_age,y)
```

#### R2 y mean squared error

```python
## R2
r2_score(y_true=y, y_pred=model_age.predict(X_age))
```

```python
## Mean Squared error
mean_squared_error(y_true=y, y_pred=model_age.predict(X_age))
```

#### Ecuación de la regresión

```python
print(model_age.intercept_, model_age.coef_)
```

$$ Precio = 177411.79 + 530.75 * Housing  Median  Age$$

#### Gráfica de las predicciones

```python
predicciones_age = model_age.predict(X_age) 
df=df.copy()
df['predicciones']=predicciones_age
plt.scatter(df.housing_median_age, df.median_house_value)
plt.plot(df.housing_median_age, df.predicciones, color='red')
```

#### Analisis Gráfica

Podemos ver que la linea recta que expresa las predicciones hechas con base al modelo casi no tiene una pendiente, indicandonos que no logra captar gran variabilidad en los datos, cuando se puede ver con los puntos azules que si existe una gran dispersión en los datos recolectados. Este modelo es insuficiente y no se puede llegar a una buena predicción así.
